```python
import os
import cv2
from skimage.measure import label, regionprops
import numpy as np

# Diretório de entrada para as imagens com os maiores contornos
input_dir = './originalGrayDatasetsegmtion/maiores_contornos'
output_dir = './originalGrayDatasetsegmtion/props'

# Certifique-se de que o diretório de saída existe
os.makedirs(output_dir, exist_ok=True)

# Função para processar uma imagem e obter as propriedades das regiões conectadas
def process_image(image_path, output_path):
    # Carregar a imagem com os contornos em escala de cinza
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Erro ao carregar a imagem {image_path}")
        return

    # Binarizar a imagem (assumindo que os contornos estão em branco)
    _, binary_image = cv2.threshold(image, 1, 255, cv2.THRESH_BINARY)

    # Rotular as regiões conectadas
    label_img = label(binary_image, connectivity=2)

    # Obter as propriedades das regiões conectadas
    props = regionprops(label_img)

    # Salvar as propriedades em um arquivo de texto
    with open(output_path, 'w') as file:
        for prop in props:
            file.write(f"Label: {prop.label}\n")
            file.write(f"Area: {prop.area}\n")
            file.write(f"Bounding Box: {prop.bbox}\n")
            file.write(f"Centroid: {prop.centroid}\n")
            file.write(f"Orientation: {prop.orientation}\n")
            file.write("\n")

    print(f"Propriedades salvas em {output_path}")

# Processar todas as imagens no diretório de contornos
for image_name in os.listdir(input_dir):
    input_path = os.path.join(input_dir, image_name)
    output_path = os.path.join(output_dir, f"{os.path.splitext(image_name)[0]}_props.txt")
    process_image(input_path, output_path)

print("Processamento concluído.")
```
